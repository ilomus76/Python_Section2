# 데이터프레임에서 원하는 데이터를 추룰하기 

```python
import pandas as pd
```

```python
data = [['김민재', 27, 75, 5428000],
['이강인', 22, 57, 3428000],
['박찬호', 50, 91, 8428000],
['차범근', 70, 80, 4428000],
['추신수', 43, 100, 4528000],
['손흥민', 31, 72, 7028000],
['황희찬', 28, 69, 2528000]]
df = pd.DataFrame(data, columns=['성명', '나이', '몸무게', '급여'])
df
```

```python
#열(column) 데이터 추출
names = df['성명'] # Series 자료형으로 결과를 줌
names
```

### [분석 연습]

```python
# 축구 선수들의 급여 평균, 야구선수들의 급여 평균
football_df=df.iloc[[0,1,3,5,6],:] # : - 시작 부터 끝까지
football_df['급여'].mean()
```

```python
baseball_df=df.iloc[[2,4],:]
baseball_df['급여'].mean()
```

```python
#조건에 맞는 데이터만 추출해보기
```

```python
#급여가 5000000원 이상인 고객연봉자 추출--불리언 인덱싱
df['급여']>5000000
```

```python
#booleam 값을 가진 시지르의 요소중에서  true인 요소만 선택하여 새로운 데이터프렘임 생성
```

```python
high_salary_df=df.loc[df['급여']>5000000]
high_salary_df
```

```python
df[df['급여']>5000000]  
```

```python
#조건이 2개 이상인 것들... 급여가 500만 이상이고, 나이가 40세 미만
(df['급여']>5000000)&(df['나이']<40)  # 비트연산자
```

```python
#조건에 맞는 사람의 셩명,나이,급여 만 추출하여 새로운 데이터프레임 만들기 
young_rich_df = df.loc[(df['급여']>5000000)&(df['나이']<40), ['성명','나이','급여']]
young_rich_df
```

```python
#or 조건... 둘중에 하나라도 만족하면..
#나이가 40이상이거나.. 몸무게가 80kg 이상인 사람들 '내시경 필수 검진 대상자'

(df['나이']>=40)|(df['몸무게']>=80)   # 비트연산자 |
```

```python
#내시경 필수 검진 대상자 추출
checkup_df = df.loc[(df['나이']>=40)|(df['몸무게']>=80)]
checkup_df
```

```python
#NOT(~) 급여가 평균 이상이 아닌 경우(평균 미만인 경우)
~(df['급여']>=df['급여'].mean())
```

```python
#평균 급여에 미치지 못하는 선수들 추출 
low_salary= df.loc[~(df['급여']>=df['급여'].mean())]
low_salary
```

```python
#특정조건에 해당할때 새로운 컬룸에 식별글씨 저장하기
df.loc[df['몸무게']>90,['체질량']]='비만'
df
```
