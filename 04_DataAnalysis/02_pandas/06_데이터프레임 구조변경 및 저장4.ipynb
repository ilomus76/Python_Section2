## 데이터 프레임 구조 변경 및 저장
1. DataFrame에 새로운 컬룸을 추가하거나 원하는 자료형으로 변환하는 등의 조작.
2. DataFrame 객체를 csv, excel 등의 파일로 저장 및 열기

```python
# 데이터 프레임 만들기
import pandas as pd

data = {
    '이름' : ['손흥민', '김민재', '이강인', '황의찬', '류현진', '김병현', '박찬호', '추신수'],
    '학교' : ['축구고', '축구고', '축구고', '축구고', '야구고', '야구고', '야구고', '야구고'],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90]
}

df= pd.DataFrame(data)
df
```

```python
#인덱스 변경해보기
df.index=['1번','2번','3번','4번','5번','6번','7번','8번']
df
```

```python
#인덱스 컬룸에 컬룸명을 설정하기
df.index.name='학생번호'
df
```

```python
#굳이 인덱스명를 학생번호로 안해도 될것같아..그래서 원래대로 0,1,2...인덱스를 초기화
df.reset_index() 
```

```python
# reset_index()를 하면 기존 인덱스로 쓰면 컴룸들이 데이터용 컬룸으로 변경됨
# 만약. 이 학생번호가 데이터로서 필요없다면..reset_index()할때 기존 인덱스용 컴룸들을 제거하도록
df.reset_index(drop=True)
```

```python
# 위 리셋기능작업을 수행하면 원본 데이터프레임은 그대로 이고.
# 변경된 새로운 데이터프레임이 출력되었던 것임
# 그래서 현재 데이터프레임을 출력해보면.. 위 reset_index()작업이 반영되어 있지 않음
df
```

```python
# reset_index()를 수행할때 원본 데이터프레임에 적용되게 하려면..
df.reset_index(drop=True, inplace=True)
df
```

```python
# 새로운 인덱스를 지정할건데..특정 컬룸의 값으로 인덱스명으로 사용하도록 변경
df.set_index('이름', inplace=True)
df
```

```python
#인덱스명을 기준으로 오름차순 또는 내림차순으로 정렬가능
df.sort_index() #inplace를 사용하지 않으면 원본은 그대로.
```

```python
df.sort_index(ascending=False)
```

```python
# 인덱스를 다시 지정하여 원위치...
df.reset_index(inplace=True)
df
```

### 새로운 컬룸 추가

```python
# 데이터셋에 평균점수를 저장되어 있지 않음. 분석할때 마다 평균계산하지 말고..학생별 평균컬룸을 만들기
# 데이터프레임 자체에 통으로 평균계산 가능
# df.mean() #error
```

```python
# 데이터프레임의 컬룸들 중에서 숫자데이터만 평균을 계산
df[['국어','영어','수학']].mean()
```

```python
# 학생별 평균을 구하기 위해 평균계산을 가로방향으로 하도록..
df[['국어','영어','수학']].mean(axis=1)
```

```python
# 평균값 시리즈를 원래 데이터프레임의 '평균'칸을 새로 만들어 추가
df['평균']= df[['국어','영어','수학']].mean(axis=1)
df
```

```python
#평균을 소수점 2자리까지만 표기
df['평균']= df[['국어','영어','수학']].mean(axis=1).round(2)
df
```

```python
# '총점'컬룸 추가하기
df['총점']= df[['국어','영어','수학']].sum(axis='columns') #axis=1
df
```

```python
# 컬룸 제거
df.drop(['총점'], axis=1)
```

```python
df
```

```python
#inplace를 사용하지 않고 기능의 결과를 저장하려면..
df= df.drop(['총점'], axis=1)
df
```

```python
# 데이터 분석을 위하여 열심히 데이터구조를 변경하면서 작업한 결과물을..
# 다음날 되면 모두 잃어버림.. 그럼 이 작업을 또 수행...이건 짜증...
# 그래서 최종 데이터프레임을 파일로 저장..
```

## Pickle 파일 저장하기 
> 파이썬 객체(값 여러가지 가지고 있는 것 , padas, series 도 객체) 를 그대로 저장해주는 표준모듈-설치할 필요없다.

문자열을 저장하는 일반적인 파일입출력과 다르게 피클파일은 객체를 그대로(binary 데이터)로 저장할 수 있어서 AI, ML 의 학습된 모델(특성값,가중치(곡선의 기울기)) 을 저장하는 용도로 많이 사용됨

```python
#1. 피클로 객체 저장하기( dump )
import pickle 

# 저장할 파이썬의 데이터 - 어떤 타입이든 저장 
person = {'name':'sam','age':20,'hobby':['swimming','running']}

# 피클파일에 딕셔너리를 통으로 저장하기 
# with open('./save_files/person.pickle)
with open('./save_files/person.pkl','wb') as file:
# with open('./save_files/person.p)
    pickle.dump(person,file)
```

```python
#2.저장된 피클파일 읽어오기 (load)
with open('./save_files/person.pkl','rb') as file:
    p = pickle.load(file)

#불러온 데이터 출력
print(p['name'],p['age'],p['hobby'])
```

```python
#판다스의 데이터프레임도 그대로 피클파일로 저장하고 열기
with open('./save_files/score.pickle','wb') as file:
    pickle.dump(df,file)
```
