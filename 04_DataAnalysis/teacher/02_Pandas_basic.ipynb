### 데이터를 Series 나 DataFrame 객체로 만드는 이유
시리즈나 데이터프레임을 사용하면 pandas 의 기초통계 기능(평균,최대,최소 등)들을 쉽게 사용할 수 있음.

#### 시리즈 사용 사례
예) 파이썬 교과목의 중간고사 점수 분포 분석

```python
scores= pd.Series(
    [78, 85, 62, 90, 55, 88],
    index=['손흥민','김민재','박찬호','김연아','박세리','김동현']
)
scores
```

```python
#시리즈를 이용한 데이터 분석 맛보기
#기본 통계
print('평균:', scores.mean() )
print('최고점:', scores.max() )
print('최저점:', scores.min() )
```

```python
#조건 필터링 (80점 이상) -- 불리언 인덱싱
passed= scores[ scores>=80 ]
passed
```

#### 데이터프레임 사용 사례
예) 학생별 중간 / 기말 점수 성적 비교 및 종합 분석

```python
#학생 성적 데이터
data = {
    "중간고사": [78, 85, 62, 90, 55, 88],
    "기말고사": [80, 88, 70, 92, 60, 90]
}

df = pd.DataFrame(
    data,
    index=["손흥민", "김민재", "박찬호", "김연아", "박세리", "김동현"]
)

df
```

```python
# 데이터프레임 기본 분석 맛보기

# 고사별 평균 확인
print('고사별 평균:')
print( df.mean() ) #데이터프레임의 전체 평균을 구하면 컬룸별 평균을 계산해줌.
```

```python
#학생별 평균을 구해서 새로운 컬룸에 저장하기
df['평균']= df.mean(axis=1) #축 - 평균을 가로방향으로 계산
print('학생별 평균이 추가된 데이터프레임')
df
```

```python
#조건부 필터링 + 정렬

# 평균 80점 이상인 고득점 학생만 선택하여 새로운 데이터프레임으로 생성
high_scores= df[ df['평균']>=80 ]
high_scores
```

```python
#우수 학생 중 평균 점수 기준 내림차순 정렬
result= high_scores.sort_values(by='평균', ascending=False)
result
```
