# 데이터프레임에서 원하는 데이터를 추출하기

```python
import pandas as pd
```

```python
data = [['김민재', 27, 75, 5428000],
        ['이강인', 22, 57, 3428000],
        ['박찬호', 50, 91, 8428000],
        ['차범근', 70, 80, 4428000],
        ['추신수', 43, 100, 4528000],
        ['손흥민', 31, 72, 7028000],
        ['황희찬', 28, 69, 2528000]]
df = pd.DataFrame(data, columns=['성명', '나이', '몸무게', '급여'])
df
```

```python
# 열(column) 데이터 추출
names= df['성명'] # Series 자료형으로 결과를 줌
names 
```

```python
type(names)
```

```python
#[]인덱싱 대신 .컬룸명 으로 추출도 가능함
df.성명
```

```python
# []안에 리스트 형태로 컬룸을 지정하면 결과를 DataFrame으로 줌
name_df= df[['성명']]
name_df
```

```python
type(name_df)
```

```python
#성명, 급여 컬룸을 한번에 추출
name_salary_df= df[ ['성명','급여'] ]
name_salary_df
```

```python
#모양 확인
name_salary_df.shape
```

```python
#각 컬룸의 자료형 확인
name_salary_df.dtypes
```

```python
# 행 데이터 주출
# 1. loc : index명 으로 추출 location
# 2. iloc : index 번호(행 위치 0,1,2,...)로 추출 integer location
```

```python
df
```

```python
#첫번째 데이터 추출
df.loc[0] #인덱스 명 -- 결과 Series로 줌
```

```python
#결과를 데이터프레임으로...
df.loc[[0]]
```

```python
#'차범근' 의 데이터를 추출
df.loc[[3]]
```

```python
# 김민재, 이강인, 손흥민, 황희찬 데이터 추출
df.loc[ [0, 1, 5, 6] ]
```

```python
#인덱스 번호와 명이 혼동되니 인덱스명을 새로 지정하기
df.index= ['a','b','c','d','e','f','g']
df
```

```python
df.loc[[0]] #error
```

```python
df.loc[['a']]
```

```python
# 인덱스번호(행 위치번호) 로 추출 iloc
df
```

```python
#김민재 데이터를 행번호로 찾기
df.iloc[0]
```

```python
df.iloc[[0]]
```

```python
#박찬호 추신수 황희찬 데이터 추출
df.iloc[[2,4,6]]
```

```python
#마지막 행의 데이터 추출
df.iloc[[-1]]
```

```python
df.iloc[[-3]]
```

```python
# 이강인 부터 추신수까지 데이터를 모두 추출 -- 슬라이스..할때는 무조건 []가 1개
df.iloc[1:5] #1~4
```

```python
#인덱스명으로 슬라이싱
df['b':'e'] #시작명, 끝명 (마지막 인덱스명까지 포함)
```

```python
# 행,열 모두를 지정하여 선택하기
# loc[[행],[열]]
# iloc[[행번호],[열번호]]
```

```python
# 'a'행의 '성명' 데이터만 추출
df.loc[['a'],['성명']]
```

```python
# c행의 성명,나이 데이터만 추출
df.loc[['c'],['성명','나이']]
```

```python
# 이강인 추신수 손흥민의 나이와 급여 추출
df.loc[['b','e','f'],['나이','급여']]
```

```python
# 이강인부터 추신수까지 성명,급여 추출 -- 슬라이싱
df.loc['b':'e',['성명','급여']]
```

```python
# 처음부터 차범근 까지 성명 나이 급여 추출
df.loc[:'d',['성명','나이','급여']]
```

```python
#처음부터 끝까지 급여만 추출
df.loc[:,['급여']]
```

```python
# iloc 사용
# 3번째 줄의 성명,급여 추출
df.iloc[[2], ['성명','급여']] # 컬룸명도 컬룸번호를 써야함.
```

```python
df.iloc[[2],[0,3]]
```

```python
#급여 컬룸의 경우는 마지막 위치임..
df.iloc[[2,5,6], [0,-1]]
```

```python
# 박찬호부터 끝 행까지... 성명부터 몸무게까지 추출
#df.iloc[2:, 0:3] # 행번호 2~, 컬룸번호 0~2
#df.iloc[2:,:3]
df.iloc[2:,:-1]
```

### [분석 연습]

```python
# 축구선수들의 급여 평균, 야구선수들의 급여 평균
football_df= df.iloc[[0,1,3,5,6],:]
football_df['급여'].mean()
```

```python
baseball_df= df.iloc[[2,4],:]
baseball_df['급여'].mean()
```

```python
#조건에 맞는 데이터만 추출해보기
```

```python
#급여가 5000000원 이상이 고액연봉자 추출 -- 불리언 인덱싱
df['급여']>5000000
```

```python
#boolean 값을 가진 시리즈의 요소중에서 true인 요소만 선택하여 새로운 데이터프레임 생성
high_salary_df= df.loc[ df['급여']>5000000 ]  
high_salary_df
```
