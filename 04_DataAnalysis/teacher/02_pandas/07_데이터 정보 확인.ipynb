## 데이터 정보 확인
데이터를 분석하려면 일단 데이터가 어떻게 생겼는지. 어떤정보들이 있는지..
자료형은 무엇인지..등 데이터를 이해하는 과정이 가장 먼저 수행됨.

```python
#데이터프레임 만들기
import pandas as pd

data = {
    '이름' : ['손흥민', '김민재', '이강인', '황의찬', '류현진', '김병현', '박찬호', '추신수'],
    '학교' : ['축구고', '축구고', '축구고', '축구고', '야구고', '야구고', '야구고', '야구고'],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90]
}
df= pd.DataFrame(data)
df
```

```python
# 계산 가능한 수치형 데이터에 대해 개수, 평균, 표준편차, 사분위값, 최대,최소 등을 한눈에 보기
df.describe()
```

```python
# 데이터의 행개수, 컬룸정보, 개수, 자료형 등 확인
df.info()
```

```python
# 데이터프레임의 값들만 확인 - 
df.values
```

```python
#인덱스 정보 확인
df.index
```

```python
#컬룸 정보
df.columns
```
