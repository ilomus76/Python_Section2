```python
import pandas as pd

df= pd.read_csv('./data/gapminder.tsv', sep='\t')
df
```

```python
# 연도별로 그룹화하여 기대수명의 평균을 구하기
#1. 데이터를 연도별로 그룹화
year_df= df.groupby('year')
year_df
```

```python
#2. 기대수명(lifeExp)칸을 추출
year_lifeExp_df= year_df['lifeExp']
year_lifeExp_df
```

```python
#3. 선택한 칸의 평균 계산
year_lifeExp_df.mean()
```

```python
#위 3단계의 작업을 보통은 한줄로 처리함.
df.groupby('year')['lifeExp'].mean()
```

```python
# 2개 이상의 열을 그룹화하기

# 각 연도별, 대륙별. 기대수명과 1인당 GDP 평균
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
```
