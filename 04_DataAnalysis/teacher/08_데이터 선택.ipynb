```python
# 특정 컬룸과 행을 선택할때 loc, iloc 말고 인덱스 [] 만으로 추출

import pandas as pd
data = {
    '이름' : ['손흥민', '김민재', '이강인', '황의찬', '류현진', '김병현', '박찬호', '추신수'],
    '학교' : ['축구고', '축구고', '축구고', '축구고', '야구고', '야구고', '야구고', '야구고'],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90]
}
df= pd.DataFrame(data)
df
```

```python
#iloc 을 사용하지 않고도 특정 컬룸의 특정행을 추출할 수 있음.
df['국어']
```

```python
df['국어'][0] #특정 행 추출 - 인덱싱
```

```python
df['국어'][2:6] #슬라이싱
```

```python
# loc, iloc 와 같은 것인가?
# 행의 인덱스명을 변경
df.index= ['a','b','c','d','e','f','g','h']
df
```

```python
df['국어'][0] #0번째 행 추출
```

```python
df['국어'].iloc[0]
```

```python
df[['국어','수학']].iloc[0]
```

```python
# loc[[행],[열]]
# iloc[[행번호],[열변호]]
df.iloc[[0],[0]]
```
