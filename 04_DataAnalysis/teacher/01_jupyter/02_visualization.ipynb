## 시각화 예제1 : 그래프
- matplotlib 를 이용하여 그래프 그려보기

```python
# anaconda 를 설치했기에 별도의 추가 설치없이도 이미 matplotlib를 설치한 상태임.

# 모듈 사용
import matplotlib.pyplot as plt
```

```python
# 그래프의 x축, y축 데이터들 준비
xs= [1,2,3,4]
ys= [1100, 2000, 2900, 3800]
```

```python
# 선 그래프 그리기
plt.plot(xs, ys)
plt.xlabel('quantity of apples')
plt.ylabel('price')
plt.show()
```

## 시각화 예제 2 : pandas table
- 행렬구조의 2차원 표구조의 데이터를 보기 좋게 보여줌.

```python
# anaconda 를 설치했기에 이미 pandas 모듈을 설치된 상태
import pandas as pd

#롯데리아 월별 매출 데이터 준비 (가상)
data = {
    '월': ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'],
    '매출액(만원)': [4200, 3900, 4500, 4700, 5200, 5800, 6100, 6400, 5900, 5500, 5300, 5000]
}

df= pd.DataFrame(data)
```

```python
#출력해보기
print(df)
```
